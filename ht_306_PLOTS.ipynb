```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Data points
x1 = np.array([3334.324938, 5792.802216, 8242.007675, 10739.72204, 3413.492264, 3352.135543, 3343.465585])
y1 = np.array([4.103426081, 5.069867399, 5.554106536, 5.97461377, 4.362081951, 4.624074288, 4.831665341])

x2 = np.array([3334.324938, 5792.802216, 8242.007675, 10739.72204, 3413.492264, 3352.135543, 3343.465585])
y2 = np.array([21.23092894, 37.84767828, 52.65932915, 66.6483811, 21.64380771, 21.3246557, 21.27906838])

# Linear fit for x1, y1 (Experimental data)
slope1, intercept1, _, _, _ = stats.linregress(x1, y1)
line1 = slope1 * x1 + intercept1

# Linear fit for x2, y2 (Theoretical data)
slope2, intercept2, _, _, _ = stats.linregress(x2, y2)
line2 = slope2 * x2 + intercept2

# Plotting
plt.figure(figsize=(8,6))

# Plot experimental data with linear fit
plt.plot(x1, y1, 'o', label='Experimental data', color='blue')
plt.plot(x1, line1, '-', color='blue')

# Plot theoretical data with linear fit
plt.plot(x2, y2, 's', label='Theoretical data', color='red')
plt.plot(x2, line2, '--', color='red')

# Labels and title
plt.xlabel('Reynolds number')
plt.ylabel('Nusselt number')
plt.title('Reynolds vs Nusselt number')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data for x and y
x1 = np.array([0.0206375, 0.034357639, 0.048084653, 0.06180125])
y1 = np.array([48.9862558, 47.41856736, 48.13908486, 48.85057274])
y2 = np.array([42.68109492, 45.32549577, 46.80149986, 47.87690057])
y3 = np.array([35.8, 40.5, 42.6, 44.1])

# Calculate linear fits for each set
slope1, intercept1, r_value1, p_value1, std_err1 = linregress(x1, y1)
slope2, intercept2, r_value2, p_value2, std_err2 = linregress(x1, y2)
slope3, intercept3, r_value3, p_value3, std_err3 = linregress(x1, y3)

# Create figure and axis
fig, ax = plt.subplots()

# Plot the data with error bars
error = 0.2  # example of a small error for each point
ax.errorbar(x1, y1, yerr=error, fmt='d', label='LMTD fit', color='blue')
ax.errorbar(x1, y2, yerr=error, fmt='o', label='ε-NTU fit', color='orange')
ax.errorbar(x1, y3, yerr=error, fmt='s', label='Experimental fit', color='green')

# Plot the linear fits
ax.plot(x1, slope1*x1 + intercept1, color='blue', linestyle='--', label=f'LMTD fit: y={slope1:.2f}x + {intercept1:.2f}')
ax.plot(x1, slope2*x1 + intercept2, color='orange', linestyle='--', label=f'ε-NTU fit: y={slope2:.2f}x + {intercept2:.2f}')
ax.plot(x1, slope3*x1 + intercept3, color='green', linestyle='--', label=f'Experimental fit: y={slope3:.2f}x + {intercept3:.2f}')

# Set axis limits to include the origin
ax.set_xlim(0, max(x1)*1.1)  # x-axis range from 0 to 10% beyond the max x1
ax.set_ylim(0, max(y1.max(), y2.max(), y3.max())*1.1)  # y-axis from 0 to 10% beyond the max y value

# Add labels and title
ax.set_xlabel('Flow rate of hot water')
ax.set_ylabel('Tout (°C)')
ax.set_title('Linear Fit of Tout vs Flow rate (keeping flow of cold fluid constant)')

# Show legend
ax.legend()

# Show grid
ax.grid(True)

# Show plot
plt.show()

```
