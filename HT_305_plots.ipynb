```python
import matplotlib.pyplot as plt

# X-axis values (Time in minutes)
time = [15, 30, 45, 60, 75, 90, 105, 120]

# Y1-axis values (Fin+Tube Heat Exchanger)
y1 = [104, 86, 86, 84, 65, 78, 82, 81]

# Y2-axis values (Bare Tube Heat Exchanger)
y2 = [75, 75, 67, 73]  # Only 4 points for Y2, corresponding to the first 4 time points

# Plotting the lines
plt.plot(time, y1, label="Fin+Tube Heat Exchanger", marker='o', linestyle='-', color='b')
plt.plot(time[:4], y2, label="Bare Tube Heat Exchanger", marker='x', linestyle='--', color='r')  # Use only 4 time points for y2

# Labeling the axes
plt.xlabel('Time (min)')
plt.ylabel('Volume of Condensate (mL)')

# Adding a title
plt.title('Volume of Condensate (mL) vs Time (min)')

# Adding a legend
plt.legend()

# Displaying the graph
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# X-axis values (Time in minutes)
time = [30, 60, 90, 120]

# Y1-axis values (Theoretical Efficiency)
y1 = [0.409, 0.409, 0.4314, 0.41495]

# Y2-axis values (Observed Efficiency)
y2 = [0.20168, 0.1942, 0.2065, 0.1913]

# Plotting the lines
plt.plot(time, y1, label="Theoretical", marker='o', linestyle='-', color='b')
plt.plot(time, y2, label="Observed", marker='x', linestyle='--', color='r')

# Labeling the axes
plt.xlabel('Time (min)')
plt.ylabel('Efficiency')

# Setting Y-axis to start from 0
plt.ylim(0, max(max(y1), max(y2)) + 0.05)  # Adjust the upper limit to fit the data better

# Adding a title
plt.title('Efficiency (Theoretical and Observed) vs Time (min)')

# Adding a legend
plt.legend()

# Displaying the graph
plt.grid(True)
plt.show()
```
